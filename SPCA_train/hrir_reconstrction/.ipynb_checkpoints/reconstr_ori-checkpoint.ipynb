```python
import scipy.io as scio
import numpy as np
from numpy.matlib import repmat
from scipy.fftpack import hilbert
```

```python
# load data
basis_train_front = scio.loadmat('../net data/basis_train_ori_front.mat')['basis_train_ori_front']

basis_train_back = scio.loadmat('../net data/basis_train_ori_back.mat')['basis_train_ori_back']
hav_train_front = scio.loadmat('../net data/hav_train_ori_front.mat')['hav_train_ori_front']
hav_train_back = scio.loadmat('../net data/hav_train_ori_back.mat')['hav_train_ori_back']
basis_test_front = scio.loadmat('../net data/basis_test_ori_front.mat')['basis_test_ori_front']
basis_test_back = scio.loadmat('../net data/basis_test_ori_back.mat')['basis_test_ori_back']
hav_test_front = scio.loadmat('../net data/hav_test_ori_front.mat')['hav_test_ori_front']
hav_test_back = scio.loadmat('../net data/hav_test_ori_back.mat')['hav_test_ori_back']
total_l = scio.loadmat('../net data/total_l.mat')['total_l']
total_r = scio.loadmat('../net data/total_r.mat')['total_r']


itd_test_front = scio.loadmat('../net data/itd_test_front_total.mat')['itd_test_front']
itd_test_back = scio.loadmat('../net data/itd_test_back_total.mat')['itd_test_back']
itd_train_front = scio.loadmat('../net data/itd_train_front_total.mat')['itd_train_front']
itd_train_back = scio.loadmat('../net data/itd_train_back_total.mat')['itd_train_back']
"""
itd_test_front = scio.loadmat('../py_netdata/itd_test_front_total.mat')['itd_test_front_total']
itd_test_back = scio.loadmat('../py_netdata/itd_test_back_total.mat')['itd_test_back_total']
itd_train_front = scio.loadmat('../py_netdata/itd_train_front_total.mat')['itd_train_front_total']
itd_train_back = scio.loadmat('../py_netdata/itd_train_back_total.mat')['itd_train_back_total']


itd_test_front = scio.loadmat('../py_netdata/itd_test_front.mat')['itd_test_front']
itd_test_back = scio.loadmat('../py_netdata/itd_test_back.mat')['itd_test_back']
itd_train_front = scio.loadmat('../py_netdata/itd_train_front.mat')['itd_train_front']
itd_train_back = scio.loadmat('../py_netdata/itd_train_back.mat')['itd_train_back']
"""

num = 200 # 主成分个数
train_num = 60
coe_train = np.zeros((num, 200, train_num))

# load training coeff in list of length 100
trainData = list(range(0,101))
for n in range(1,102):
    trainData[n-1] = scio.loadmat('../net data/coeff_f'+str(n)+'train_out.mat')['train_out']
    coe_train[:, n-1, :] = np.transpose(trainData[n-1])

index = int(train_num/2)
coe_train_l = coe_train[:, :, 0:index]
coe_train_r = coe_train[:, :, index:train_num]

for n in range(102,201):
    coe_train_l[:, n-1, :] = coe_train_l[:, 202-n-1, :]
    coe_train_r[:, n-1, :] = coe_train_r[:, 202-n-1, :]

test_num = 14
coe_test = np.zeros((num, 200, test_num))

# load training coeff in list of length 100
testData = list(range(0,101))
for n in range(1,102):
    testData[n-1] = scio.loadmat('../net data/coeff_f'+str(n)+'test_out.mat')['test_out']
    coe_test[:, n-1, :] = np.transpose(testData[n-1])

index = int(test_num/2)
coe_test_l = coe_test[:, :, 0:index]
coe_test_r = coe_test[:, :, index:test_num]

for n in range(102,201):
    coe_train_l[:, n-1, :] = coe_train_l[:, 202-n-1, :]
    coe_train_r[:, n-1, :] = coe_train_r[:, 202-n-1, :]
    
# 合并系数
coe_l = np.concatenate([np.transpose(coe_train_l, (2,0,1)), np.transpose((coe_test_l), (2,0,1))],axis=0)
coe_l = np.transpose(coe_l, (1,2,0))
coe_r = np.concatenate([np.transpose(coe_train_r, (2,0,1)), np.transpose(coe_test_r, (2,0,1))],axis=0)
coe_r = np.transpose(coe_r, (1,2,0))



total_num = 37

# Reconstruct HRTF

# 930个方向，1~625左耳；467~930右耳
# 除测试集外930组数据
dataSet = [1] + list(range(2,155,4)) + list(range(170,455,4)) + list(range(470,1251,4)) +\
           list(range(3,56,4)) + list(range(71,556,4)) + list(range(571,780,4)) +\
           list(range(795,1080,4)) + list(range(1095,1248,4)) + list(range(4,681,4)) +\
           list(range(696,1181,4)) + list(range(1196,1249,4)) +\
           [1237, 1241, 1245, 1249, 9, 13, 17, 309, 313, 317, 609, 613, 617, 933, 937, 941]
# 训练集：750组数据
trainSet = dataSet[1:928:5] + dataSet[2:929:5] + dataSet[3:930:5] + dataSet[4:931:5] +\
           [901, 906, 911, 916, 921, 926] 
validSet = dataSet[0 :897 : 5]
testSet = [5] + list(range(21,306,4)) + list(range(321,606,4)) + list(range(621,930,4)) +\
          list(range(945,1234,4)) +\
          [59, 63, 67, 158, 162, 166, 458, 462, 625, 559, 563, 567, 684, 688, 692, 783, 787, 791, 1083, 1087, 1091, 1184, 1188, 1192]
def minusOne(list):
    for _ in range(0,len(list)):
        list[_] -= 1

minusOne(testSet)
minusOne(dataSet)
minusOne(trainSet)
minusOne(validSet)

trainingdata = trainSet + validSet
totalSet = trainingdata + testSet

# get sorted index of list totalSet, like [~, inds] = sort(totalSet) in matlab does
sortedTotal = sorted(totalSet)
inds = [totalSet.index(sortedTotal[0])]
for i in range(1,len(totalSet)):
    inds += [totalSet.index(sortedTotal[i])]
    
# 合并基函数
basis_front = np.concatenate([basis_train_front, basis_test_front],axis=0)
basis_back = np.concatenate([basis_train_back, basis_test_back],axis=0)
hav_front = np.concatenate([hav_train_front, hav_test_front],axis=0)
hav_back = np.concatenate([hav_train_back, hav_test_back],axis=0)

# 调整基函数排序为 1~1250
basis_front = basis_front[inds, :] # 左耳水平角-80～80,右耳80～-80（不同坐标系？），下同
basis_back = basis_back[inds, :]
hav_front = hav_front[inds, :]
hav_back = hav_back[inds, :]

hd_l_front = np.zeros((total_num, 200, 625))
hd_r_front = np.zeros((total_num, 200, 625))
hrtf_l_front = np.zeros((total_num, 200, 625))
hrtf_r_front = np.zeros((total_num, 200, 625))
hd_l_back = np.zeros((total_num, 200, 625))
hd_r_back = np.zeros((total_num, 200, 625))
hrtf_l_back = np.zeros((total_num, 200, 625))
hrtf_r_back = np.zeros((total_num, 200, 625))

for i in range(0,total_num):
    for l in range(0,625):
        hd_l_front[i, :, l] = np.transpose(coe_l[:, :, i])@np.transpose(basis_front[l, :])
        hd_l_back[i, :, l] = np.transpose(coe_l[:, :, i])@np.transpose(basis_back[l, :])
        hrtf_l_front[i, :, l] = hd_l_front[i, :, l] + hav_front[l] + total_l
        hrtf_l_back[i, :, l] = hd_l_back[i, :, l] + hav_back[l] + total_l
        r = l + 625 # 1250 - ceil(l/25)*25 + mod(l-1, 25) + 1; # 调整右耳水平角顺序为-80～80
        hd_r_front[i, :, l] = np.transpose(coe_r[:, :, i])@np.transpose(basis_front[r, :])
        hd_r_back[i, :, l] = np.transpose(coe_r[:, :, i])@np.transpose(basis_back[r, :])
        hrtf_r_front[i, :, l] = hd_r_front[i, :, l] + hav_front[r] + total_r
        hrtf_r_back[i, :, l] = hd_r_back[i, :, l] + hav_back[r] + total_r

```

```python
def itd_add(hrir, fs, itd):
    z = np.fft.fft(hrir,200)
    mag = np.abs(z)
    ang = np.unwrap(np.angle(z))
    f = np.transpose(np.linspace(0,z.shape[0],z.shape[0])) * fs / z.shape[0]
    ang[0:100] = (ang[0:100]-2 * np.pi * itd * f[0:100])
    z[0:100] = mag[0:100] * np.cos(ang[0:100]) + (0+1.j) * mag[0:100] * np.sin(ang[0:100])
    re,im = np.real(z), np.imag(z)
    re[100], im[100] = 0,0
    im[0] = 0
    for i in range(100,200): # 实部偶对称，虚部奇对称
        re[i]= re[100-(i-100)]
        im[i]= -im[100-(i-100)]
    tf=re+(0+1.j)*im
    hrir[:] = np.real(np.fft.ifft(tf,200))
    return hrir

def reconstr(hrtf_l, hrtf_r, itd, direc, pn):
    # 信号重建
    hrir_l, hrir_r = np.zeros((pn,200)), np.zeros((pn,200))
    rebuiltphase_l = np.zeros((pn, 200))
    magn_l = np.power(10,hrtf_l/20) # 左耳幅度
    magn_l_log = np.log(magn_l); # 左耳对数幅度，以e为底
    

    for i in range(0,pn):
        rebuiltphase_l[i, :] = -np.imag(hilbert(magn_l_log[i, :])) #重建相位
    hr_l = magn_l*np.cos(rebuiltphase_l)+(0+1.j)*magn_l*np.sin(rebuiltphase_l) #重建后hrtf的对数表示
    for i in range(0,pn):
        hrir_l[i, :] = np.real(np.fft.ifft(hr_l[i, :])) #左耳时域
    rebuiltphase_r = np.zeros((pn, 200))
    magn_r = np.power(10,hrtf_r/20) # 左耳幅度
    magn_r_log = np.log(magn_r) # 左耳对数幅度，以e为底
    for i in range(0,pn):
        rebuiltphase_r[i, :] = -np.imag(hilbert(magn_r_log[i, :])) #重建相位
    hr_r = magn_r*np.cos(rebuiltphase_r)+(0+1.j)*magn_r*np.sin(rebuiltphase_r) # 重建后hrtf的对数表示
    for i in range(0,pn):
        hrir_r[i, :] = np.real(np.fft.ifft(hr_r[i, :])) # 右耳时域
    fs = 44100
    itd = itd/fs # 转化成时间单位（秒）

    if direc > 625:
        for i in range(0,itd.shape[0]):
            hrir_l[i, :] = itd_add(hrir_l[i, :], fs, itd[i]);
    elif direc < 625:
        for i in range(0,itd.shape[0]):
            hrir_r[i, :] = itd_add(hrir_r[i, :], fs, itd[i])
    elif direc == 625:
        for i in range(0,itd.shape[0]):
            if np.max(hrir_l[i, :]) > np.max(hrir_r[i, :]):
                hrir_r[i, :] = itd_add(hrir_r[i, :], fs, itd[i])
            else:
                hrir_l[i, :] = itd_add(hrir_l[i, :], fs, itd[i])
    else:
        raise IndexError('Wrong direction input')
    return hrir_l, hrir_r

# 重建hrir
itd_front = np.concatenate([itd_train_front, itd_test_front],axis=0)
itd_back = np.concatenate([itd_train_back, itd_test_back],axis=0)
"""
itd_front = itd_front.reshape(625,total_num)
itd_back = itd_back.reshape(625,total_num)
"""
"""
itd_front = repmat(itd_front,1,37)
itd_back = repmat(itd_back,1,37)
"""
itd_front = itd_front.reshape(625, total_num) # 方向先变仰角后水平角，301～325为中平面
itd_back = itd_back.reshape(625, total_num)

hrir_front_l = np.zeros((total_num, 200, 625))
hrir_front_r = np.zeros((total_num, 200, 625))
hrir_back_l = np.zeros((total_num, 200, 625))
hrir_back_r = np.zeros((total_num, 200, 625))

for i in range(0,625):
    print('Spatial directions: ',i)
    dirc = (np.ceil(i/25)*50-25) # 调整301～325为625
    hrir_front_l[:, :, i], hrir_front_r[:, :, i] = reconstr(hrtf_l_front[:, :, i], hrtf_r_front[:, :, i], itd_front[i, :], dirc, total_num)
    hrir_back_l[:, :, i], hrir_back_r[:, :, i] = reconstr(hrtf_l_back[:, :, i], hrtf_r_back[:, :, i], itd_back[i, :], dirc, total_num)
    
```
