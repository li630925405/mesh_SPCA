```python
def itd_add(hrir, fs, itd):
    z = np.fft.fft(hrir,200)
    mag = np.abs(z)
    ang = np.unwrap(np.angle(z))
    f = np.transpose(np.linspace(0,z.shape[0],z.shape[0])) * fs / z.shape[0]
    ang[0:100] = (ang[0:100]-2 * np.pi * itd * f[0:100])
    z[0:100] = mag[0:100] * np.cos(ang[0:100]) + (0+1.j) * mag[0:100] * np.sin(ang[0:100])
    re,im = np.real(z), np.imag(z)
    re[100], im[100] = 0,0
    im[0] = 0
    for i in range(100,200): # 实部偶对称，虚部奇对称
        re[i]= re[100-(i-100)]
        im[i]= -im[100-(i-100)]
    tf=re+(0+1.j)*im
    hrir[:] = np.real(np.fft.ifft(tf,200))
    return hrir

def reconstr(hrtf_l, hrtf_r, itd, direc, pn):
    # 信号重建
    hrir_l, hrir_r = np.zeros((pn,200)), np.zeros((pn,200))
    rebuiltphase_l = np.zeros((pn, 200))
    magn_l = np.power(10,hrtf_l/20) # 左耳幅度
    magn_l_log = np.log(magn_l); # 左耳对数幅度，以e为底
    

    for i in range(0,pn):
        rebuiltphase_l[i, :] = -np.imag(hilbert(magn_l_log[i, :])) #重建相位
    hr_l = magn_l*np.cos(rebuiltphase_l)+(0+1.j)*magn_l*np.sin(rebuiltphase_l) #重建后hrtf的对数表示
    for i in range(0,pn):
        hrir_l[i, :] = np.real(np.fft.ifft(hr_l[i, :])) #左耳时域
    rebuiltphase_r = np.zeros((pn, 200))
    magn_r = np.power(10,hrtf_r/20) # 左耳幅度
    magn_r_log = np.log(magn_r) # 左耳对数幅度，以e为底
    for i in range(0,pn):
        rebuiltphase_r[i, :] = -np.imag(hilbert(magn_r_log[i, :])) #重建相位
    hr_r = magn_r*np.cos(rebuiltphase_r)+(0+1.j)*magn_r*np.sin(rebuiltphase_r) # 重建后hrtf的对数表示
    for i in range(0,pn):
        hrir_r[i, :] = np.real(np.fft.ifft(hr_r[i, :])) # 右耳时域
    fs = 44100
    itd = itd/fs # 转化成时间单位（秒）

    if direc > 625:
        for i in range(0,itd.shape[0]):
            hrir_l[i, :] = itd_add(hrir_l[i, :], fs, itd[i]);
    elif direc < 625:
        for i in range(0,itd.shape[0]):
            hrir_r[i, :] = itd_add(hrir_r[i, :], fs, itd[i])
    elif direc == 625:
        for i in range(0,itd.shape[0]):
            if np.max(hrir_l[i, :]) > np.max(hrir_r[i, :]):
                hrir_r[i, :] = itd_add(hrir_r[i, :], fs, itd[i])
            else:
                hrir_l[i, :] = itd_add(hrir_l[i, :], fs, itd[i])
    else:
        raise IndexError('Wrong direction input')
    return hrir_l, hrir_r

# 重建hrir
itd_front = np.concatenate([itd_train_front, itd_test_front],axis=0)
itd_back = np.concatenate([itd_train_back, itd_test_back],axis=0)
"""
itd_front = itd_front.reshape(625,total_num)
itd_back = itd_back.reshape(625,total_num)
"""
"""
itd_front = repmat(itd_front,1,37)
itd_back = repmat(itd_back,1,37)
"""
itd_front = itd_front.reshape(625, total_num) # 方向先变仰角后水平角，301～325为中平面
itd_back = itd_back.reshape(625, total_num)

hrir_front_l = np.zeros((total_num, 200, 625))
hrir_front_r = np.zeros((total_num, 200, 625))
hrir_back_l = np.zeros((total_num, 200, 625))
hrir_back_r = np.zeros((total_num, 200, 625))

for i in range(0,625):
    print('Spatial Directions: ',i)
    dirc = (np.ceil(i/25)*50-25) # 调整301～325为625
    hrir_front_l[:, :, i], hrir_front_r[:, :, i] = reconstr(hrtf_l_front[:, :, i], hrtf_r_front[:, :, i], itd_front[i, :], dirc, total_num)
    hrir_back_l[:, :, i], hrir_back_r[:, :, i] = reconstr(hrtf_l_back[:, :, i], hrtf_r_back[:, :, i], itd_back[i, :], dirc, total_num)
    
```
